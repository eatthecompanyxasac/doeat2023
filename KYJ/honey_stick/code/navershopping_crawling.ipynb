함수설정

```python
def page_count(num, per_page):
    if num % per_page ==0:
        page = int(num / per_page)
    else:
        page = int(num / per_page) + 1
    return page
```

```python
def review_count():
    review_num = driver.find_elements(By.CLASS_NAME, f'_2pgHN-ntx6')[0].text
    if ',' in review_num:
        review_num = int(review_num.replace(',', ''))
    else:
        review_num = int(review_num)
    return review_num
```
