# 상황/먹은사람 카운팅

```python
def kw_in_product(product_name, keyword):
    return data_kurly[(data_kurly['ProductName'] == product_name)&(data_kurly['ReviewText'].apply(lambda x:keyword in x))]['ReviewText']
```

```python
invite = ['집들이', '초대', '접대', '파티', '해드', '파티', '크리스마스', '대접', '생일', '생신', '손님',
          '모실때', '모시', '어르신', '어루신']
```

```python
def kw_search_list(kw_list, sen):
    label = 0
    for kw in kw_list:    
        if kw in sen:
            label = 1
            break
    return label
```

```python
sum(data_kurly['ReviewText'].apply(lambda x:kw_search_list(invite, x)))
```

```python
data_kurly['reception'] = data_kurly['ReviewText'].apply(lambda x:kw_search_list(invite, x))
```

```python
data_kurly['challenge'] = data_kurly['ReviewText'].apply(lambda x:kw_search_list(goodmeal, x))
```

```python
data_kurly['convinence'] = data_kurly['ReviewText'].apply(lambda x:kw_search_list(easycook, x))
```

```python
kurly_when = data_kurly[(data_kurly['convinence']==1)|(data_kurly['challenge']==1)|(data_kurly['reception']==1)]
```

```python
kurly_when
```

```python
# 4: 1&2번 집단
# 5: 1&3번 집단
# 6: 2&3번 집단
# 7: 1&2&3번 집단
def kw_search_list(kw_list1, kw_list2, kw_list3, sen):
    label = 0
    for kw in kw_list1:    
        if kw in sen:
            label = 1
            break
    for kw in kw_list2:
        if kw in sen:
            if label==1:
                label=4
            else:
                label=2
            break
    for kw in kw_list3:
        if kw in sen:
            if label==4:
                label=7
            elif label==1:
                label=5
            elif label==2:
                label=6
            else:
                label=3
            break
    return label
```

```python
data_kurly['when_label'] = (data_kurly['ReviewText'].apply(lambda x:kw_search_list(invite, goodmeal, easycook, x)))
```

```python
data_kurly[(data_kurly['reception']==0)&(data_kurly['challenge']==0)&(data_kurly['convinence']==0)]
```

```python
data_kurly[data_kurly['when_label']==0]
```
