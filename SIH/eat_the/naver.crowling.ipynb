```python
import requests
from bs4 import BeautifulSoup

def crawl_naver_cafe_blog(keyword):
    url = f"https://search.naver.com/search.naver?sm=tab_hty.top&where=view&query=%22{keyword}%22&oquery={keyword}&tqi=i6R5qsprvTVssUlhOXRssssstKK-300658&mode=normal"

    # HTTP GET 요청
    response = requests.get(url)
    response.raise_for_status()  # 오류 발생 시 예외 처리

    # BeautifulSoup을 사용하여 HTML 파싱
    soup = BeautifulSoup(response.text, 'html.parser')

    # 크롤링 결과를 담을 리스트
    results = []

    # 검색 결과를 추출하는 부분
    items = soup.select('.api_txt_lines.dsc_txt')
    for item in items:
        result = item.get_text(strip=True)
        results.append(result)

    return results

if __name__ == "__main__":
    keyword = "솥밥 밀키트"
    contents = crawl_naver_cafe_blog(keyword)
    for idx, content in enumerate(contents, start=1):
        print(f"{idx}. {content}")
```
