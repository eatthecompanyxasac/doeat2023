```python
import pandas as pd
import numpy as np
```

```python
naver_nouns = pd.read_csv(r'C:\Users\NT550\asac\SIH\SIH\code2\code\shopping_nouns.csv')
naver_nouns.head()
```

```python
import ast
# 불용어 리스트
stop_words = ['ㅎ','포','ㅋ','ㅋㅋ','ㅎㅎ','ㅠ','ㅠㅠ','ㅜ','ㅜㅜ','달','거','것','어용','앞']

# 불용어 제거 함수
def remove_stop_words(tokens):
    tokens_list = ast.literal_eval(tokens)  # 문자열 형태의 리스트를 실제 리스트로 변환
    filtered_tokens = [token for token in tokens_list if token.strip() not in stop_words]
    return filtered_tokens

# 'cleaned_tokens' 열에 있는 값에 불용어 제거 함수를 적용하여 새로운 열 생성
naver_nouns['cleaned_tokens'] = naver_nouns['token_nouns'].apply(remove_stop_words)


naver_nouns.head()
```

```python
naver_nouns.to_csv('naver_shopping_preprocess.csv')
```
