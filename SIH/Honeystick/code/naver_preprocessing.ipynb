### 불용어 처리

```python
import ast


# 불용어 리스트
stop_words = ['꿀','스틱','사용기','뭐','달','수','날','데','저','이번','ㅎ','포','ㅋ','ㅋㅋ','ㅎㅎ','ㅠ','ㅠㅠ','ㅜ','ㅜㅜ','달',
              '거','것','꺼','어용','앞','저희','나','저','점','전','포','후','이', '있', '하', '것', '들', '그', '되', '수', '이',
              '보','데', '듣','제','분','이거','번','손''않', '없', '나', '사람', '주', '아니', '등', '같', '우리', '때', '년', '가', '한', '지', '대하', '오', '말', '일', '그렇', '위하','으','잘','시','ㅡ','만']

# 불용어 제거 함수
def remove_stop_words(tokens):
    tokens_list = ast.literal_eval(tokens)  # 문자열 형태의 리스트를 실제 리스트로 변환
    filtered_tokens = [token for token in tokens_list if token.strip() not in stop_words]
    return filtered_tokens

# 'cleaned_tokens' 열에 있는 값에 불용어 제거 함수를 적용하여 새로운 열 생성
naver_nouns['cleaned_tokens'] = naver_nouns['token_nouns'].apply(remove_stop_words)


naver_nouns.head()
```

```python
naver_nouns.to_csv('naver_shopping_preprocess.csv')
```

```python
naver_shopping = pd.read_csv(r'C:\Users\NT550\asac\SIH\SIH\code2\result\naver_shopping_preprocess.csv')
naver_shopping.head()
```

```python
### 감성분석
# import library
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TextClassificationPipeline

# load model
tokenizer = AutoTokenizer.from_pretrained("jaehyeong/koelectra-base-v3-generalized-sentiment-analysis")
model = AutoModelForSequenceClassification.from_pretrained("jaehyeong/koelectra-base-v3-generalized-sentiment-analysis")
sentiment_classifier = TextClassificationPipeline(tokenizer=tokenizer, model=model)
```

```python
naver_shopping['리뷰'].dropna(inplace=True)
```

```python
naver_a = naver_shopping[55173:55229]
```

```python
review_list = list(naver_a['리뷰'])
```

```python
# 감정분석 결과 및 점수
pred_label = []
pred_score = []
for rev in review_list:
    label_temp = sentiment_classifier(rev)[0]['label']
    score_temp = sentiment_classifier(rev)[0]['score']
    pred_label.append(label_temp)
    pred_score.append(score_temp)
```

```python
naver_a['sentiment_label'] = pred_label
naver_a['sentiment_score'] = pred_score
```

```python
naver_a.loc[naver_a['sentiment_label'] == '1', 'sentiment_label'] = '긍정'
naver_senti = naver_a['sentiment_label']
```

```python
# 'sentiment_label' 값을 가진 행 삭제
naver_senti = naver_senti[naver_senti != 'sentiment_label']
```
