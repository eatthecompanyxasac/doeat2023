### 블로그 별로 명사 동사 형용사 나눠서 데이터프레임에 추가

#### 명사

##### 명사만 토큰화한 것을 데이터 프레임에 추가

```python
# 명사 추출 및 짧은 단어 제외하는 함수
def process_text(x):
    okt = Okt()
    nouns = okt.nouns(x)
    nouns_okt = [n for n in nouns if len(n) > 1]
    return nouns_okt

# 데이터프레임의 각 행에 함수 적용
naver_blog_text_scaling['명사 토큰화'] = naver_blog_text_scaling['변환된 내용'].apply(lambda x : process_text(str(x)))
```

##### 동사 토큰화 컬럼 추가

```python
# 명사 추출 및 짧은 단어 제외하는 함수
def process_text(x):
    komoran = Komoran()
    verbs_komoran = komoran.pos(x, flatten=True)
    verbs_komoran = [word for word, pos in verbs_komoran if pos.startswith('VV') and len(word) > 1]
    return verbs_komoran

# 데이터프레임의 각 행에 함수 적용
naver_blog_text_scaling['동사 토큰화'] = naver_blog_text_scaling['변환된 내용'].apply(lambda x : process_text(str(x)))
```

##### 형용사 토큰화 컬럼 추가

```python
# 명사 추출 및 짧은 단어 제외하는 함수
def process_text(x):
    komoran = Komoran()
    adjectives_komoran = komoran.pos(x) # 명사만 추출
    adjectives_komoran = [word for word, pos in adjectives_komoran if pos.startswith('VA') and len(word) > 1]
    return adjectives_komoran

# 데이터프레임의 각 행에 함수 적용
naver_blog_text_scaling['형용사 토큰화'] = naver_blog_text_scaling['변환된 내용'].apply(lambda x : process_text(str(x)))
```

```python
naver_blog_text_scaling["변환된 내용"] = naver_blog_text_scaling["변환된 내용"].apply( lambda x : str(x))
naver_blog_text_scaling
```

##### 토큰화 컬럼들에 내가 고른 키워드를 검색하여 주소와 내용보는 코드

```python
# 특정 문자를 찾는 함수
def find_list_with_word(x, target_word, row):
    if x == "명사":
        return target_word in row['명사 토큰화']
    elif x == "동사":
        return target_word in row['동사 토큰화']
    else:
        return target_word in row['형용사 토큰화']

# find_blog_url_with_word 함수 수정
def find_blog_url_with_word(x, target_word, row):
    rows_with_target_word = naver_blog_text_scaling[naver_blog_text_scaling.apply(lambda row: find_list_with_word(x, target_word, row), axis=1)]
    selected_columns = ['주소']  # 원하는 열을 선택할 경우
    rows_with_selected_columns = rows_with_target_word[selected_columns]
    return rows_with_selected_columns
```

```python
def extract_context(text, keyword, num_characters=20):
    pattern = re.compile(rf'(?P<context>.*?(?<=\b{re.escape(keyword)}\b).{{0,{num_characters}}}(?=\b{re.escape(keyword)}\b).*)', re.IGNORECASE)
    match = re.search(pattern, text)
    if match:
        return match.group('context').strip()
    return ""
# 결과를 저장할 리스트 초기화
result = []

# 각 행에 대해 키워드를 찾아서 앞뒤 20글자를 추출하고 결과를 리스트에 저장
for index, row in naver_blog_text_scaling.iterrows():
    context = extract_context(row["변환된 내용"], "압력", 20)
    result.append(context)

# 결과 리스트 출력
print(result)
```
